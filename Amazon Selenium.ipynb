```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
```

```python
driverPath =  r"C:\\Users\\ELCOT\driver\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverPath)

driver.get("https://www.amazon.in/")
driver.maximize_window()
writer=pd.ExcelWriter(r"C:\\Users\\ELCOT"+"\\"+'amazon.xlsx',engine='xlsxwriter')
driver.find_element_by_xpath("//a[contains(text(),'Fashion')]").click()
driver.find_element_by_xpath("//body/div[@id='a-page']/div[2]/div[2]/div[1]/div[1]/section[1]/div[1]/div[1]/div[1]/div[1]/div[1]/ol[1]/li[1]/a[1]").click()
driver.find_element_by_xpath("//a[@id='a-autoid-1-announce']").click()

clothing=[]
prod=driver.find_elements_by_xpath(r"//div[contains(@class,'a-row a-size-base a-color-base')]")
for i in prod:
    clothing.append(i.text.split())
    print(clothing)
    



    
pd.ExcelWriter('amazon.xlsx',engine='xlsxwriter')
t=pd.DataFrame(clothing,columns=["OLD PRICE","NEW PRICE","SAVINGS","SAVERUPEES","DISCOUNT",""])
t.to_excel(writer,sheet_name="data",index=False)

print(driver.title)
print(driver.current_url)

writer.save()
writer.close()
```

```python
driverPath =  r"C:\\Users\\ELCOT\driver\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverPath)
driver.get("https://www.amazon.in/s?k=hard+disk&i=computers&rh=n%3A1375395031%2Cp_n_pct-off-with-tax%3A2665399031&dc&qid=1630143814&rnid=2665398031&ref=sr_nr_p_n_pct-off-with-tax_1")
name = driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
originalprice = driver.find_elements_by_xpath("//span[@class='a-price a-text-price']")
offerprice = driver.find_elements_by_xpath("//span[@class='a-price']")
lst = []
lst1 = []
lst2 = []

writer = pd.ExcelWriter(r"C:\Users\Merit\Desktop\Python_Assignment-main\amazon.xlsx", engine='xlsxwriter')
for i in name:
    lst.append(i.text)

for j in originalprice:
    lst1.append(j.text)

for k in offerprice:
    lst2.append(k.text)


df = pd.DataFrame(list(zip(lst,lst1,lst2)), columns = ['Product Name','Original Price','Offer Price'])
df.to_excel(writer,index = False)
writer.save()
writer.close()
print(df)
driver.close()
```
