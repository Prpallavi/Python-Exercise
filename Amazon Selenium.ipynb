```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
```

```python
driverPath =  r"C:\\Users\\ELCOT\driver\chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverPath)

driver.get("https://www.amazon.in/")
driver.maximize_window()
writer=pd.ExcelWriter(r"C:\\Users\\ELCOT"+"\\"+'amazon.xlsx',engine='xlsxwriter')
driver.find_element_by_xpath("//a[contains(text(),'Fashion')]").click()
driver.find_element_by_xpath("//body/div[@id='a-page']/div[2]/div[2]/div[1]/div[1]/section[1]/div[1]/div[1]/div[1]/div[1]/div[1]/ol[1]/li[1]/a[1]").click()
driver.find_element_by_xpath("//a[@id='a-autoid-1-announce']").click()

clothing=[]
prod=driver.find_elements_by_xpath(r"//div[contains(@class,'a-row a-size-base a-color-base')]")
for i in prod:
    clothing.append(i.text.split())
    print(clothing)
    



    
pd.ExcelWriter('amazon.xlsx',engine='xlsxwriter')
t=pd.DataFrame(clothing,columns=["OLD PRICE","NEW PRICE","SAVINGS","SAVERUPEES","DISCOUNT",""])
t.to_excel(writer,sheet_name="data",index=False)

print(driver.title)
print(driver.current_url)

writer.save()
writer.close()
```
